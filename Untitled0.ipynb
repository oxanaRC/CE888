```
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)

# Now create a figure
plt.figure()
# and plot x inside it
plt.plot(x)
```

```
# Now create a new figure
plt.figure()
# and plot y inside it...
plt.plot(y)
# Add a title
plt.title("Cos")
```
